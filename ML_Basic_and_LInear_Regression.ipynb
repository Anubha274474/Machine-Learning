```python
# Question8:- Select Independent variables (features) and dependent variable (target).
# target variable : medv (house price)
X=df.drop('medv',axis=1) #Independent variables
y=df['medv'] # Dependent variables
```

```python
# Question9:- Split the dataset into training and testing sets (80% tarining,20% testing).
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
# Question10:- Train a Linear Regression model and perform the following:
#  Make prediction
#  Calculate MAE
#  Calculate MSE
# Calculate RMSE
# Calculate R square
# Calculate Adjusted R square 
# Write the Linear Regression Equation
# Interpret the model coefficients
```

```python
# Model Train + Evaluate
model= LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
# Metrics calculate karo
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2=r2_score(y_test,y_pred)
```

```python
# Adjusted R square
n = X_test.shape[0] #samples
p= X_test.shape[1] # features
adj_r2= 1-(1-r2)*(n-1)/(n-p-1)
```

```python
print('Mean_absolute_error:',mae)
```

```python
print('Mean_squared_error:',mse)
```

```python
print('Square_root_Mean_squared_error:',rmse)
```

```python
print('R square:',r2)
```

```python
print('Adjusted R square:',adj_r2)
```

```python
# Regression Coefficient 

print('\nRegression Equation:')
print(f'medv={model.intercept_:.2f}',end='')
for i, col in enumerate (X.columns):
    print(f'+({model.coef_[i]:2f}*{col}',end='')
    
```

```python
# Coefficient Interpretation
coef_df =pd.DataFrame({'Feature':X.columns,'Coefficient': model.coef_})
coef_df=coef_df.sort_values('Coefficient',ascending=False)
print(coef_df)
```
